***Zadanie 6.***  
Dla głównej funkcji oraz przedziału zdefiniowanego z zadania 1 znajdź miejsca zerowe przy użyciu:

* metody bisekcji
* metody siecznych
* metody Newtona

z tolerancją $10^{-10}$

Zbadaj dokładność(względem rozwiązania z zadania 5) i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach. 

```python
a1 = -0.8
b1 = 0.3
a2 = 0.5  # Przedziały potrzebne do znalezienia poszczególnych miejsc zerowych
b2 = 3

epsilon = 10e-10 #Żądana dokładność
max_iter = [1, 3, 5, 10, 14,18,20,25,30,35,40,50,60] #Liczba iteracji
```

```python
#Metoda bijekcji

first_root_norm = []
second_root_norm = []

for i in max_iter:
    first_root = main.bisection(a1, b1, main.fun, epsilon, i)
    second_root = main.bisection(a2, b2, main.fun, epsilon, i)
    first_root_norm.append(np.abs(real_root[0] - first_root[0]))
    second_root_norm.append(np.abs(real_root[1] - second_root[0]))

plt.figure()
plt.scatter(max_iter , first_root_norm, color = 'magenta')
plt.grid()
plt.title('Znalezienie pierwszego pierwiastka  za pomocą metody bisekcji')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

plt.figure()
plt.scatter(max_iter, second_root_norm, color = 'cyan')
plt.grid()
plt.title('Znalezienie drugiego pierwiastka za pomocą metody bisekcji')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()


for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.bisection(a1, b1, main.fun, epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'magenta')
    plt.title('Czas obliczeń pierwszego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()
plt.show()

plt.figure()

for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.bisection(a2, b2, main.fun, epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'cyan')
    plt.title('Czas obliczeń drugiego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()    


```

```python
#Metoda siecznych


first_root_norm = []
second_root_norm = []

for i in max_iter:
    first_root = main.secant(a1, b1, main.fun, epsilon, i)
    second_root = main.secant(a2, b2, main.fun, epsilon, i)
    first_root_norm.append(np.abs(real_root[0] - first_root[0]))
    second_root_norm.append(np.abs(real_root[1] - second_root[0]))

plt.figure()
plt.scatter(max_iter , first_root_norm, color = 'magenta')
plt.grid()
plt.title('Znalezienie pierwszego pierwiastka za pomocą metody siecznych')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

plt.figure()
plt.scatter(max_iter, second_root_norm, color = 'cyan')
plt.grid()
plt.title('Znalezienie drugiego pierwiastka za pomocą metody siecznych')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()


for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.secant(a1, b1, main.fun, epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'magenta')
    plt.title('Czas obliczeń pierwszego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()
plt.show()

plt.figure()

for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.secant(a2, b2, main.fun, epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'cyan')
    plt.title('Czas obliczeń drugiego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()    
```

```python
#Metoda Newtona


first_root_norm = []
second_root_norm = []

for i in max_iter:
    first_root = main.newton(main.fun,main.dfun,main.ddfun,a1,b1, epsilon, i)
    second_root = main.newton(main.fun,main.dfun,main.ddfun,a2,b2 ,epsilon, i)
    first_root_norm.append(np.abs(real_root[0] - first_root[0]))
    second_root_norm.append(np.abs(real_root[1] - second_root[0]))

plt.figure()
plt.scatter(max_iter , first_root_norm, color = 'magenta')
plt.grid()
plt.title('Znalezienie pierwszego pierwiastka za pomocą metody Newtona')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

plt.figure()
plt.scatter(max_iter, second_root_norm, color = 'cyan')
plt.grid()
plt.title('Znalezienie drugiego pierwiastka za pomocą metody Newtona')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()


for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.newton(main.fun,main.dfun,main.ddfun,a1,b1, epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'magenta')
    plt.title('Czas obliczeń pierwszego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()
plt.show()

plt.figure()

for i in max_iter:
    method_time = %timeit -r 10 -n 10 -o main.newton(main.fun,main.dfun,main.ddfun,a2,b2 ,epsilon, i)
    
    plt.scatter(i, method_time.average, color = 'cyan')
    plt.title('Czas obliczeń drugiego pierwiastka w zależności od ilości iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.semilogy()    
```
